```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
```

```python
label_encoders = {}
for column in df.select_dtypes(include=['object']).columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])

# Display encoding information
for i in list(label_encoders.keys()):
    classes = label_encoders[i].classes_
    encoding_dict = {class_label: index for index, class_label in enumerate(classes)}
    print(i, encoding_dict)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
for feature in ['Vehicle_Speed', 'Crash_Time', 'Age', 'Lane_Width', 'Speed_Limit']:
  df[feature] = scaler.fit_transform(df[[feature]])
```

```python
import seaborn as sns
sns.pairplot(df,hue="Crash_Severity")
```

```python
# Define features and target
X = df.drop('Crash_Severity', axis=1)
y = df['Crash_Severity']
```

```python
# Scale selected numerical features
scaler = MinMaxScaler()
X_scaled = X.copy()  # Copy original X to preserve structure
numerical_features = ['Vehicle_Speed', 'Crash_Time', 'Age', 'Lane_Width', 'Speed_Limit']
X_scaled[numerical_features] = scaler.fit_transform(X[numerical_features])
```

```python
# Apply PCA for dimensionality reduction
pca = PCA(n_components=0.95)  # Retain 95% of variance
X_pca = pca.fit_transform(X_scaled)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.1, random_state=42, shuffle=True)
```

```python
# Train KNN classifier
knn = KNeighborsClassifier(n_neighbors=4, weights='uniform')
knn.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
error_rate=[]
for i in range(2,30):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(range(,30),error_rate,color='blue',linestyle='dashed',marker='o')
```
